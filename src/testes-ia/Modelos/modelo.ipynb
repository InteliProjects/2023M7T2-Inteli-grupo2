```python
import pandas as pd
```

```python
df = pd.read_parquet('tabelaPHGeral.parquet')
```

```python
df
```

```python
#create a column that increase by 1 and reset when the value of the column 'Aircraft' changes
df['indexAircraft'] = df.groupby('Aircraft').cumcount() + 1
```

```python
df
```

```python
#chage the values of the column aircraft to values that icrease by 1 each time the value of the column 'Aircraft' changes
df['Aircraft'] = df.groupby('Aircraft').ngroup() + 1
df
```

```python
df.head(48)
```

```python
#find the values of Flight in each aircraft
t = df.groupby('Aircraft')['Flight'].unique()
for i in t:
    x = 0
    for n in i:
        x = x + 1
        #change n to a value
        df.loc[df['Flight'] == n, 'Flight'] = x
        print(n, x)

df
```

```python
df.groupby('Aircraft')['Flight'].unique()
```

```python
df
```

```python
df = df.sort_values(by=['Aircraft','Flight','phaseOfFlight-1'])
df
```

```python
#create new column 'preFail' that is 1 in the 5 flights of each aircraft that precede the failure and 0 in the others
df['preFail'] = 0
df
```

```python
#find the values different of 0 in the columns message0418DAA-1 and message0422DAA-1; in the 15 rows before the value found, preFail is 1
for i in df.index:
    if df.loc[i,'message0418DAA-1'] != 0 or df.loc[i,'message0422DAA-1'] != 0:
        for n in range(1,16):
            print(i, n)
            if i - n >= 0:
                df.loc[i - n,'preFail'] = 1
```

```python
#show the rows where preFail is 1 and the rows where message0418DAA-1	message0422DAA-1 are different of 0
df.loc[df['preFail'] == 1]
```

```python
#remove message0418DAA-1	message0422DAA-1 from df
df = df.drop(['message0418DAA-1','message0422DAA-1'], axis=1)
```

```python
df.head(48)
```

```python
#normalize "Duration" column
df['Duration'] = df['Duration']/df['Duration'].max()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
```

```python
df.head(48)
```

```python
df = df.sort_values(by=['Flight','phaseOfFlight-1','indexAircraft'])
df.head(48)
```

```python
df
```

```python
df_x = df.drop('preFail', axis = 'columns')
df_y = df['preFail']
```

```python
df_x
```

```python
df_y
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=5)
pca_res = pca.fit_transform(df_x)
```

```python
df_x = pd.DataFrame(pca_res)
```

```python
df_x
```

```python
df_y
```

```python
X = df_x
Y = df_y
x_train, x_test, y_train, y_test = train_test_split(X,Y,test_size = 0.3, random_state = 2)
```

```python
clf = RandomForestClassifier(n_estimators = 100)
clf.fit(x_train, y_train)
y_pred = clf.predict(x_test)
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(clf,X,Y,cv=5)
```

```python
scores
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, accuracy_score
y_true = y_test
cm = confusion_matrix(y_true, y_pred)
cm
```

```python

disp = ConfusionMatrixDisplay(confusion_matrix=cm)

disp.plot()
```

```python

print("ACCURACY OF THE MODEL: ",accuracy_score(y_test, y_pred))
```
