Colocando data em formato de data no df

```python
#create column date in format datetime based on dateDay-1 dateMonth-1 dateYear-1

#drop NaN values
df = df.dropna()

df['dateDay-1'] = df['dateDay-1'].astype(int)
df['dateMonth-1'] = df['dateMonth-1'].astype(int)
df['dateYear-1'] = df['dateYear-1'].astype(int)
df['date'] = (df['dateYear-1'].astype(str) + '-' + df['dateMonth-1'].astype(str) + '-' + df['dateDay-1'].astype(str))
df = df.drop(['dateDay-1', 'dateMonth-1', 'dateYear-1'], axis=1)
df
```

```python
#drop row where date = '0-1-1', '16-1-14', '4-1-2', '128-1-2'
for i in ['0-1-1','16-1-14','4-1-2','128-1-2']:
    df = df[df['date'] != i]
    
df['date'].unique()
```

```python
AmM = AmM.sort_values(by=['Aircraft', 'date'])
AmM
```

```python
AmM['Aircraft'] = AmM.groupby('Aircraft').ngroup() + 1
AmM
```

```python
tsm = AmM[['Aircraft', 'M']].copy()
tsm
```

```python
tsm['tsm'] = AmM['Duration'].copy()
```

```python
tsm
```

```python
tsm = tsm.fillna(0)
tsm
```

```python
tsmp = pd.DataFrame(columns=tsm.columns)
for i in tsm['Aircraft'].unique():
    atsm = tsm[tsm['Aircraft'] == i].copy()
    atsm['tsm'] = atsm['tsm'].groupby(atsm['M']).cumsum()
    tsmp = pd.concat([tsmp, atsm], axis=0)

tsmp
```
