```python
caminho_da_pasta = "c:/Users/rafak/Documents/a1dataAzul"  

lista_arquivos = os.listdir(caminho_da_pasta)
print(lista_arquivos)
```

```python
lista_arquivos = lista_arquivos[0:16]
print(lista_arquivos)
```

```python
countAviao= 0
relatorioPHgeral = pd.DataFrame(columns = ['Aircraft','Flight','phaseOfFlight-1', 'message0418DAA-1', 'message0422DAA-1','Duration','bleedMonPress-1b','bleedMonPress-2b','bleedOutTemp-1b', 'bleedOutTemp-2b','dateDay-1', 'dateMonth-1', 'dateYear-1'])
for i in lista_arquivos:
    countAviao += 1
    countVoo = 0
    try:
        lista_voos = os.listdir(caminho_da_pasta+'/'+i)
        for n in lista_voos:
            countVoo += 1
            voo = pd.read_parquet(f'{caminho_da_pasta}/{i}/{n}')
            voo = voo[['phaseOfFlight-1', 'message0418DAA-1', 'message0422DAA-1','recording_time','dateDay-1', 'dateMonth-1', 'dateYear-1','bleedOutTemp-1b', 'bleedOutTemp-2b','bleedMonPress-1b','bleedMonPress-2b']]
            voo = voo.fillna(method='ffill')
            messages = voo.groupby(['phaseOfFlight-1']).max()
            date = voo.groupby(['phaseOfFlight-1']).min()
            duration = voo.groupby(['phaseOfFlight-1']).max() - voo.groupby(['phaseOfFlight-1']).min()
            cVar = voo.groupby(['phaseOfFlight-1']).mean()
            messages = messages[['message0418DAA-1', 'message0422DAA-1']]
            duration = duration[['recording_time']]
            date = date[['dateDay-1', 'dateMonth-1', 'dateYear-1']]
            cVar = cVar[['bleedOutTemp-1b', 'bleedOutTemp-2b','bleedMonPress-1b','bleedMonPress-2b']]
            voo = pd.concat([messages, duration, date, cVar], axis=1)
            voo = voo.rename(columns={'recording_time': 'Duration'})
            voo['phaseOfFlight-1'] = voo.index
            voo = voo.reset_index(drop=True)
            voo['Flight'] = n[-22:-8]
            voo['Aircraft'] = i
            relatorioPHgeral = pd.concat([relatorioPHgeral, voo], axis=0, ignore_index=True)
            print(f"Avião {countAviao} de {len(lista_arquivos)}; voo {countVoo} de {len(lista_voos)}")
    except:
        print(i,n)
```
