```python
nData = pd.read_parquet("C:/Users/rafak/Documents/a1dataAzul/06120083/TCRF_ARCHIVE_06120083_20221226053232.parquet")
```

```python

    
nData = nData[['aircraftSerNum-1', 'phaseOfFlight-1', 'message0418DAA-1', 'message0422DAA-1','recording_time','dateDay-1', 'dateMonth-1', 'dateYear-1','bleedOutTemp-1b', 'bleedOutTemp-2b','bleedMonPress-1b','bleedMonPress-2b']]
nData = nData.fillna(method='ffill')
messages = nData.groupby(['phaseOfFlight-1']).max()
date = nData.groupby(['phaseOfFlight-1']).min()
duration = nData.groupby(['phaseOfFlight-1']).max() - nData.groupby(['phaseOfFlight-1']).min()
cVar = nData.groupby(['phaseOfFlight-1']).mean()
messages = messages[['message0418DAA-1', 'message0422DAA-1', 'aircraftSerNum-1']]
duration = duration[['recording_time']]
date = date[['dateDay-1', 'dateMonth-1', 'dateYear-1']]
cVar = cVar[['bleedOutTemp-1b', 'bleedOutTemp-2b','bleedMonPress-1b','bleedMonPress-2b']]
nData = pd.concat([messages, duration, date, cVar], axis=1)
nData = nData.rename(columns={'recording_time': 'Duration'})
nData['phaseOfFlight-1'] = nData.index
nData = nData.reset_index(drop=True)
nData = nData.dropna()

nData['dateDay-1'] = nData['dateDay-1'].astype(int)
nData['dateMonth-1'] = nData['dateMonth-1'].astype(int)
nData['dateYear-1'] = nData['dateYear-1'].astype(int)
nData['date'] = (nData['dateYear-1'].astype(str) + '-' + nData['dateMonth-1'].astype(str) + '-' + nData['dateDay-1'].astype(str))
nData = nData.drop(['dateDay-1', 'dateMonth-1', 'dateYear-1'], axis=1)
nData['date'] = pd.to_datetime(nData['date'], format='%Y-%m-%d')
```
