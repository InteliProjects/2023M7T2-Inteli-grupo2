### Junção dos arquivos de voo de uma aeronave

O notebook apresenta a junção dos múltiplos arquivos de voo de uma única aeronave. O resultado é um arquivo de extensão .parquet com todos os voos da aeronave, isto é, um dataframe unificado com as informações disponibilizadas para cada aeronave da companhia aérea.

```python
import pandas as pd
import os
```

```python
flight = 0
folder = "./data"
archives = [arq for arq in os.listdir(folder) if arq.endswith('.parquet')]
```

Abaixo, encontra-se a lógica utilizada para a junção dos arquivos de voo da aeronave, desconsiderando quaisquer tratamentos de dados.

```python
for archive in archives:
    path_archive = os.path.join(folder, archive)
    df = pd.read_parquet(path_archive)
    df['flight'] = flight
    if flight == 0:
        df_final = df
    else:
        df_final = pd.concat([df_final, df])
    flight += 1
    df = None

df_final
```
